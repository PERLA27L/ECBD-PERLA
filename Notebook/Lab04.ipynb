```python
#Limpieza
#Nombres de columnas en minisculas
df.columns = df.columns.str.strip().str.lower()
```

```python
df.columns
```

```python
#Detectar valores nulos
df.isnull().sum()
```

```python
#Eliminar una columna cabin
df.drop(columns=["cabin"], inplace=True)
```

```python
df.columns
```

```python
#Rellenamos los datos nulos con la mediana de la columna
df["age"] = df["age"].fillna(df["age"].median())
```

```python
df.isnull().sum()
```

```python
#Rellenamos el Fare con el valor mas repetido (moda)
df["fare"] = df["fare"].fillna(df["fare"].mode()[0])
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
#Validar datos
df["age"].min()
```

```python
#Graficamos el total de pasajeros por clase
plt.figure(figsize=(8,4)) #Tamaño
sns.countplot(df, x="pclass")
plt.title("Total de pasajeros por clase")
plt.xlabel("Clase")
plt.show()
```

```python
#Graficamos las clases por genero
plt.figure(figsize=(8,4))
sns.countplot(df,  x="sex", hue="pclass")
plt.title("Total de Genero por clases")
plt.xlabel("Genero")
plt.ylabel("Total")
plt.show()
```

```python
#Mostramos la edad por genero y clase
plt.figure(figsize=(6,4))
sns.violinplot(df, x="sex", y="age", hue="pclass")
plt.title("Rango de edad por genero y clase")
plt.xlabel("Genero")
plt.ylabel("Edad")
plt.show()
```

```python
#CONOCIMIENTO
#Los pasajeros de primera clase pagaban tarifas mas altas.
df.groupby("pclass")["fare"].mean()
```
