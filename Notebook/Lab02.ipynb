```python
import pandas as pd
import random

# 1. Definición de listas (Basado exactamente en tu imagen)
cantidad = 1000000
influencers = ["LuisitoComunica", "Yuya", "JuanpaZurita", "KimLoaiza", "IbaiLlanos", "AurunPlay", "Kenini"]
tipos_post = ["foto", "video", "reel", "carrusel"]
fechas = ["2026-01-10", "2026-02-15", "2025-05-20", "2024-12-01", "2026-03-10", "2025-11-11"]

# 2. Creación del DataFrame (Misma estructura que tu práctica anterior)
dtMillion = pd.DataFrame({
    "influencer": [random.choice(influencers) for _ in range(cantidad)],
    "likes": [random.randint(5000, 1000000) for _ in range(cantidad)],
    "comentarios": [random.randint(100, 50000) for _ in range(cantidad)],
    "fecha": [random.choice(fechas) for _ in range(cantidad)],
    "tipo_post": [random.choice(tipos_post) for _ in range(cantidad)]
})

# 3. Diagnóstico de los datos (Lo que usaste en tu otra práctica)
print(dtMillion.head(5))
print("\n--- Tipos de datos ---")
print(dtMillion.dtypes)
print("\n--- Uso de memoria ---")
print(dtMillion.memory_usage(deep=True))

# 4. Análisis de Engagement (Para cumplir el objetivo de la imagen)
dtMillion["engagement"] = dtMillion["likes"] + dtMillion["comentarios"]
print("\n--- Insights de Engagement ---")
print(dtMillion.groupby("influencer")["engagement"].mean())
```
