```python
cantidad = 1000000
```

```python
import random
```

```python
ciudades = ["CDMX","Monterrey","Guadalajara","Puebla","Cancun","Merida","Tijuana","Cordoba"]
```

```python
nombres = ["Ana","Luis","Marta","Carolina","Sofia","Pedro","Elena","Diego"]
categorias = ["Tecnología","Ropa","Hogar","Videojuegos","Deportes","Calzado","Finanzas","Politica"]
```

```python
dtMillion = pd.DataFrame ({
     "nombre" : [random.choice(nombres) for _ in range(cantidad)],
    "edad" : [random.randint(18,70)for _ in range(cantidad)],
    "salario" : [round(random.uniform(5000,50000),2) for _ in range(cantidad)],
    "activo" : [random.choice([True,False]) for _ in range(cantidad)],
    "categoria" : [random.choice(categorias)for _ in range(cantidad)],
    "ciudad" : [random.choice(ciudades)for _ in range(cantidad)],
    "anio_registro": [random.randint(2020,2026)for _ in range(cantidad)]
})
```
